```python
def web_link(link):
  with urllib.request.urlopen(link) as url:
    data = json.loads(url.read().decode())
    return menu(data)
```

```python
def menu(data):
  df = pd.DataFrame.from_dict(data)
  result = pd.DataFrame({'ProductPrice':[],'CategoryId':[],'CategoryName':[],'ProductId':[],'ProductName':[],'SDMID':[],'AddOns':[]})
  for row in range(len(df)):
    ProductPrice = []
    CategoryId = []
    CategoryName = []
    ProductId = []
    ProductName = []
    SDMID = []
    AddOns = []
    df1 = pd.DataFrame.from_dict(df['categories'][row])
    for j in range(len(df1)):
      df2 = pd.DataFrame.from_dict(df1['products'][j],orient='index').transpose()
      if (len(df2['bundleProductOptions'][0])>2):
        df3 = pd.DataFrame.from_dict(df2['bundleProductOptions'][0][1])
        AO = []
        if (len(df3)!=0):
          for i in range(len(df3)):
            df4 = pd.DataFrame.from_dict(df3['productLinks'][i],orient='index').transpose()
            AO.append(df4['name'][0])
      else:
        AO = [0] 
      ProductPrice.append(df2['finalPrice'][0])
      CategoryId.append(str(df1['id'][0]))
      CategoryName.append(df1['categoryCode'][0])
      ProductId.append(str(df2['id'][0]))
      ProductName.append(df2['name'][0])
      SDMID.append(str(df2['sdmId'][0]))
      AddOns.append(AO)
    R = pd.DataFrame(list(zip(ProductPrice, CategoryId, CategoryName, ProductId, ProductName, SDMID,AddOns)), columns = ['ProductPrice','CategoryId','CategoryName','ProductId','ProductName','SDMID','AddOns'])
    result = result.append(R, ignore_index = True)
  return result
```
